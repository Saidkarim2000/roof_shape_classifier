```python
import math

def coord_to_tile(coord:list):
  zoom = 18
  lat_rad = math.radians(coord[0])
  n = 2.0 ** zoom
  xtile = int((coord[1] + 180.0) / 360.0 * n)
  ytile = int((1.0 - math.asinh(math.tan(lat_rad)) / math.pi) / 2.0 * n)
  return (xtile, ytile)
```

```python
def centroid(vertexes):
     lon_list = [vertex[0] for vertex in vertexes]
     lat_list = [vertex[1] for vertex in vertexes]
     _len = len(vertexes)
     longitude = sum(lon_list) / _len
     latitude = sum(lat_list) / _len
     return [latitude, longitude]
```

```python
geometry = []

for i in hip_roof['geometry'].astype(str):

  start = i.find("MULTIPOLYGON (((") + len("MULTIPOLYGON (((")
  end = i.find(")))")
  substring = i[start:end]
  geometry.append(substring)
  
geometry
```

```python
len(geometry)
```

```python
coords = []
for i in geometry:
  result_1 = i.split(',')
  coords.append(result_1)

coords[:10]
```

```python
len(coords)
```

```python
coords_sep = []

for i in range(len(coords)):
  coords_item = []
  for j in coords[i]:
    result_2 = j.split()
    result_2[0] = float(result_2[0].replace("(", ""))
    result_2[1] = float(result_2[1].replace(")", ""))
    coords_item.append(result_2)
  coords_sep.append(coords_item)

coords_sep
```

```python
type(coords_sep[0][1][1])
```

```python
hip_coord_center = []
for i in range(len(coords_sep)):
    center = centroid(coords_sep[i])
    hip_coord_center.append(center)

hip_coord_center
```
